# **Números y strings**

En primer lugar, Python puede manejar enteros en distintas bases, números con coma flotante y números complejos.

```python
a = 8      
b = 8.0     
c = 1.4e80 
d = 0b100   
e = 43.2 + 3j
```

Podemos verificar el tipo de datos usando type()

```python
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```
